```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import yfinance as yf
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import string
from itertools import chain
from IPython.display import Image, display
```

**For ticker extraction** https://stockmarketmba.com/stocksinthedjia.php


```python
# Request URL and Parsing the HTML content using BeautifulSoup
Ticker_URL = "https://stockmarketmba.com/stocksinthedjia.php"
request = requests.get(Ticker_URL)
soup = BeautifulSoup(request.text)
# Extracting the values
ticker = soup.find_all("tr")
# Creating the list of tickers
ticker_list = []
for i in ticker:
    if i.find("a") is not None:
        ticker_list.append(i.find("a").text)
# Cleaning the list of tickers
ticker_list = sorted(ticker_list)


ticker_list.append("^DJI")
len(ticker_list)
```

```python
yf.pdr_override()
total_set = pd.DataFrame()
df1 = pd.DataFrame()

df = pd.DataFrame()
for i in ticker_list:
    symbol = i
    start = datetime(2020,9,1)
    total_set = pdr.get_data_yahoo(symbol,start=start)
    df[symbol+'_close'] = total_set['Adj Close'] #Adjusted close is the closing price after adjustments for all applicable splits and dividend distributions
    if i == '^DJI':
        symbol = i
        start = datetime(1984,1,1)
        total_set = pdr.get_data_yahoo(symbol,start=start)
        df1 = total_set
```
